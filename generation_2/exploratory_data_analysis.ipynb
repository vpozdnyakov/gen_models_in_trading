# Exploratory data analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
```

```python
returns = pd.read_csv('data/returns.csv', index_col=0)
returns.index = pd.to_datetime(returns.index)
```

```python
fig = plt.figure(figsize=(13, 5))
for col in returns.columns:
    plt.plot(returns[col].iloc[150:200], alpha=0.5, label=col)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d %b %Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=10))

plt.xlabel('')
plt.ylabel('returns')
plt.legend()
plt.savefig('figs/returns.png')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.heatmap(
    returns.corr(), 
    vmax=0.9,
    vmin=-0.5,
    cbar=False,
    annot=True, 
    cmap=plt.cm.Reds_r)
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('figs/heatmap.pdf')
plt.show()
```

```python
sns.pairplot(
    returns, 
    x_vars=["heating", "gas", 'natural'],
    y_vars=["brent", "crude", 'gasoline'],
    plot_kws=dict(s=5, linewidth=0)
)
plt.show()
```
